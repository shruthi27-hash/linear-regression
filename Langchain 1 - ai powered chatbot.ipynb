```python
import google.generativeai as genai
import gradio as gr

# ðŸ”‘ Enter your Gemini API key here
genai.configure(api_key="YOUR_API_KEY_HERE")

# Initialize Gemini Pro Model
model = genai.GenerativeModel("gemini-pro")

# Chat history setup
chat = model.start_chat(history=[])

# Chat function
def gemini_chat(user_input):
    response = chat.send_message(user_input)
    return response.text

# Gradio Interface
gr.ChatInterface(
    fn=gemini_chat,
    title="ðŸ’¬ Gemini Chatbot",
    description="Talk with Google's Gemini using a friendly interface!",
).launch()
```

```python
import os
from langchain.chat_models import ChatOpenAI
from langchain.schema import SystemMessage, HumanMessage
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
import gradio as gr

# ðŸ”‘ Add your OpenAI API Key
os.environ["OPENAI_API_KEY"] = "your_openai_api_key_here"

# Setup LLM
llm = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0.7)

# Setup memory
memory = ConversationBufferMemory()

# Setup conversational chain
conversation = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=False
)

# Chat function
def chat_with_ai(user_input):
    return conversation.predict(input=user_input)

# Gradio interface
gr.ChatInterface(
    fn=chat_with_ai,
    title="ðŸ¤– LangChain AI Chatbot",
    description="A smart assistant powered by LangChain + OpenAI. Ask me anything!",
).launch()
```
